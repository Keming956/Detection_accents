```python
import pandas as pd
import soundfile as sf
import io
import os
```

```python

file_path = "validation-00000-of-00001-6bfe0053d478cbd2.parquet"
data = pd.read_parquet(file_path)
```

```python

output_dir = "output_audio"
os.makedirs(output_dir, exist_ok=True)
```

```python
# étiquettes d'accent
selected_accents = [
    'United States English', 
    'India and South Asia (India, Pakistan, Sri Lanka)', 
    'England English'
]

# filtrer les données
filtered_data = data[data['accent'].isin(selected_accents)]

# traiter les données
for index, row in filtered_data.iterrows():
    audio_data = row['audio']
    sentence = row['sentence']
    accent = row['accent']
    
    if 'bytes' in audio_data:
        byte_stream = io.BytesIO(audio_data['bytes'])
        waveform, sample_rate = sf.read(byte_stream)
        
        # créer un répertoire pour chaque accent
        accent_dir = os.path.join(output_dir, accent.replace(' ', '_'))
        os.makedirs(accent_dir, exist_ok=True)
        
        audio_filename = f"{accent_dir}/audio_{index}.wav"
        sf.write(audio_filename, waveform, sample_rate)
        
        # sauvegarder le texte
        text_filename = f"{accent_dir}/audio_{index}.txt"
        with open(text_filename, 'w') as text_file:
            text_file.write(sentence)
        
        print(f"Processed {index}: Saved audio and text for accent '{accent}'.")

print("Processing complete.")
```
